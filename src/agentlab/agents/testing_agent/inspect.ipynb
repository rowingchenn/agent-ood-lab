```python
import os
import json

# Calculate the number of successful experiments
def analyze_summary_files(folder_path):
    success_count = 0
    total_count = 0

    for root, dirs, files in os.walk(folder_path):
        if 'summary_info.json' in files:
            file_path = os.path.join(root, 'summary_info.json')
            with open(file_path, 'r') as f:
                try:
                    data = json.load(f)
                    terminated = data.get("terminated", False)
                    truncated = data.get("truncated", False)

                    if terminated and truncated:
                        # Print the name of the subfolder if both are True
                        print(f"Both 'terminated' and 'truncated' are true in folder: {os.path.basename(root)}")
                    elif terminated:
                        success_count += 1
                    elif truncated:
                        pass  # Count as failure but do nothing here
                    total_count += 1

                except json.JSONDecodeError:
                    print(f"Error decoding JSON in file: {file_path}")

    return success_count, total_count
```

```python
folder_path = "/home/weichen/agentlab_results/2024-11-03_16-21-32_GenericAgent-Qwen_Qwen2.5-7B-Instruct_on_workarena.l1"
success, total = analyze_summary_files(folder_path)
print(f"Success: {success}/{total}")
```
